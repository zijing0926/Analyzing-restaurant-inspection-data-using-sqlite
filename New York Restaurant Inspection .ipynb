## Remove Null entries


Return the number of inspections (`CAMIS`, `INSPDATE` pairs) that do not have a score - i.e. where none of the rows with those (`CAMIS`, `INSPDATE`) values has a score. Remove the corresponding rows from the data set for the rest of the questions in the assignment.

```python
%%sql 
SELECT * from webextract LIMIT 5;
```

```python
%%sql
with null_table as (SELECT count(SCORE) as count_null FROM webextract GROUP BY CAMIS, INSPDATE HAVING count(SCORE) =0)
select count(*)
from null_table;
```

```python
%%sql
CREATE TEMP TABLE IF NOT EXISTS FINAL_TABLE AS
SELECT * FROM WEBEXTRACT
JOIN (SELECT CAMIS, INSPDATE FROM webextract GROUP BY CAMIS, INSPDATE HAVING count(SCORE) !=0) AS N
WHERE WEBEXTRACT.CAMIS=N.CAMIS AND WEBEXTRACT.INSPDATE=N.INSPDATE;
```

```python
%%sql
SELECT * FROM FINAL_TABLE LIMIT 10
```

## Check score by zip code

Return a list of tuples of the form:

    (zip code, mean score, number of restaurants)


```python
%%sql
DROP TABLE IF EXISTS MDATE_SCORE;
CREATE TEMP TABLE IF NOT EXISTS MDATE_SCORE AS
SELECT SCORE, MAX(INSPDATE) AS MDATE,ZIPCODE,CAMIS FROM FINAL_TABLE GROUP BY ZIPCODE,CAMIS
```

```python
%%sql
CREATE TEMP TABLE IF NOT EXISTS ZIPCODE AS
SELECT ZIPCODE,AVG(SCORE),COUNT(CAMIS) 
FROM MDATE_SCORE
GROUP BY ZIPCODE
HAVING COUNT(DISTINCT CAMIS) >100
ORDER BY AVG(SCORE)
```

```python
%%sql
WITH MD AS (SELECT MAX(INSPDATE) AS MDATE,CAMIS FROM FINAL_TABLE GROUP BY CAMIS)
SELECT WEB.ZIPCODE, AVG(WEB.SCORE), COUNT(WEB.CAMIS) FROM MD
JOIN FINAL_TABLE AS WEB
WHERE MD.MDATE=WEB.INSPDATE AND MD.CAMIS=WEB.CAMIS
GROUP BY WEB.ZIPCODE HAVING COUNT(DISTINCT WEB.CAMIS)>100
ORDER BY AVG(WEB.SCORE)
```

```python
##creating restaurant*years
%%sql
DROP TABLE IF EXISTS DEN;
CREATE TEMP TABLE IF NOT EXISTS DEN AS
SELECT CAMIS, COUNT(VIOLCODE),CUISINECODE, CASE WHEN 
(MAX(JULIANDAY(INSPDATE))-MIN(JULIANDAY(INSPDATE)))>30 THEN (MAX(JULIANDAY(INSPDATE))-MIN(JULIANDAY(INSPDATE)))
ELSE 30
END AS LENGTH
FROM FINAL_TABLE
GROUP BY CAMIS
#creating ratio
%%sql
DROP TABLE IF EXISTS RATIO_TABLE;
CREATE TEMP TABLE IF NOT EXISTS RATIO_TABLE AS
WITH DEN1 AS (SELECT CUISINECODE, SUM(LENGTH*1.0/365) AS YEARS FROM DEN
GROUP BY CUISINECODE)
SELECT DEN1.CUISINECODE, COUNT(VIOLCODE)/YEARS AS RATIO
FROM FINAL_TABLE
JOIN DEN1
WHERE DEN1.CUISINECODE = FINAL_TABLE.CUISINECODE
GROUP BY DEN1.CUISINECODE
HAVING COUNT(VIOLCODE)>100
%%sql
SELECT CODEDESC, RATIO
FROM RATIO_TABLE AS A
JOIN CUISINE AS B
WHERE A.CUISINECODE = B.CUISINECODE
```
